```python
import pandas as pd
import numpy as np
```

```python
prime_unpro = pd.read_csv("data_119/all_assistance_prime_awards_1.csv")
sub_unpro = pd.read_csv("data_119/all_assistance_subawards_1.csv")
```

```python
# Selecting Relevant Columns for our Analysis
prime = prime_unpro[["assistance_award_unique_key","award_id_fain", 
        "total_funding_amount", "period_of_performance_start_date",
        "period_of_performance_current_end_date", "awarding_agency_name", 
        "award_description", "recipient_duns", "recipient_name", 
        "recipient_city_name", "recipient_county_name", 
        "business_types_description",            
        "primary_place_of_performance_city_name",
        "primary_place_of_performance_county_name",
        "primary_place_of_performance_congressional_district"]]
```

```python
prime = prime.rename(columns={"assistance_award_unique_key": "unique_key", 
                                "award_id_fain": "fain",
                                "total_funding_amount": "funding_amount",
                                "awarding_agency_name": "agency",
                                "period_of_performance_start_date": "start_date", 
                                "period_of_performance_current_end_date" : "end_date"})
```

```python
prime.head()
```

```python
prime.shape
```

```python
prime.groupby("business_types_description")["unique_key"].count()
```

```python
sub_unpro.columns
```

```python
sub = sub_unpro[["prime_award_unique_key", "prime_award_fain", "prime_award_amount",
        "prime_award_awarding_agency_name","prime_awardee_name", "prime_awardee_duns", 
        "subaward_amount", "subaward_action_date", "subawardee_duns", "subawardee_name",
        "subawardee_city_name", "subawardee_congressional_district", 
        "subaward_description"]]
```

```python
sub.shape
```

```python
full = prime.merge(sub, how="left", left_on = "unique_key", right_on = "prime_award_unique_key")
```

```python
full.head()
```

```python
prime.to_csv("data_119/prime_awards.csv", index=False)
sub.to_csv("data_119/sub_awards.csv", index=False)
full.to_csv("data_119/full_awards.csv", index=False)
```

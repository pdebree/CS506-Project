### Contract Length Analysis

```python
contracts.columns.values
```

```python
contracts.iloc[0, 2]
```

```python
end_year = [0 if str(row)[-7:-5] == '' else int(str(row)[-7:-5]) for row in contracts["period_of_performance_current_end_date"]]
end_year = [row if row < 50 else row - 100 for row in end_year]
```

```python
np.unique(end_year)
```

```python
#start_year = [0 if str(row)[-7:-5] == '' else int(str(row)[-2:]) for row in contracts["period_of_performance_start_date"]]
#start_year = [row if row < 50 else row - 100 for row in start_year]
start_year = contracts['start_year']
```

```python
awd_len = [0] * len(end_year)
for i in range(len(start_year)):
    awd_len[i] = end_year[i] - start_year[i]
#awd_len = [row if row >= 0 else 0 for row in awd_len]
awd_len
```

```python
contracts["awd_len"] = awd_len
```

```python
contracts["awd_len"]
```

```python
awd_len_b = contracts[contracts['black_american_owned_business'] == True]["awd_len"].mean()
awd_len_b
```
